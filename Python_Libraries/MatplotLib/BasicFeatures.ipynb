```python
# 1. Importing package
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 2. Basic Plot
%matplotlib inline
plt.plot([1,2,3,4,5],[2,8,5,10,3],'rD-')            #(x,y,format{color}{marker}{line})
plt.show()                              
```

```python
# 3. two basic plots
plt.plot([1,2,3,4,5],[2,4,6,8,10],'gs')
plt.plot([1,2,3,4,5],[1,4,9,16,25],'b*')
plt.show()
```

```python
# 4. Adding Title
plt.title("A Simple Plot")
plt.show()                             #a plain plot
```

```python
# 5. Adding labels
plt.plot([2,4,6,8,10],[2,4,6,8,10],'c*')
plt.plot([1,3,5,7,9],[1,4,9,16,25],'r+')
plt.title('Combined Plot')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.show()
```

```python
# 6. Legends
plt.plot([2,4,6,8,10],[2,4,6,8,10],'c*',label='cyan stars')
plt.plot([1,3,5,7,9],[1,4,9,16,25],'r+',label='red plus')
plt.title('Combined Plot')
plt.legend(loc=0)
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.show()
```
