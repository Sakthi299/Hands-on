```python
# 1. figure size of plot
from matplotlib import pyplot as plt
plt.figure(figsize=(10,5))                                                        #(width,height) 
a=plt.plot([1,2,3,4,5], [1,2,3,4,10], 'bv', label='drop')  
b=plt.plot([1,2,3,4,5], [2,3,4,5,11], 'r^', label='hike')  
plt.title('A Simple Scatterplot')  
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.legend(loc='best')
plt.show()
```

```python
# 2. fixing axis ranges
plt.plot([1,2,3,4,5],[3,6,9,12,18],'mH:')
plt.xlim(1, 6)
plt.ylim(1, 18)
plt.show()
```

```python
# 3. subplots in same figure

fig,(ax1, ax2) = plt.subplots(1,2,figsize=(10,4),dpi=200)

ax1.plot([1,2,3,4,5], [1,2,3,4,10], 'go') 
ax2.plot([1,2,3,4,5], [2,3,4,5,11], 'b*')

ax1.set_title('Scatterplot Dots'); ax2.set_title('Scatterplot Stars')
ax1.set_xlabel('X');  ax2.set_xlabel('X')
ax1.set_ylabel('Y');  ax2.set_ylabel('Y') 
ax1.set_xlim(0, 6) ;  ax2.set_xlim(0, 6)
ax1.set_ylim(0, 12);  ax2.set_ylim(0, 12)
plt.tight_layout()
plt.show()
```

```python
# 4. fixing subplots

plt.figure(figsize=(5,5), dpi=200,facecolor='lightgrey') 

plt.subplot(2,1,1)
plt.plot([1,2,3,4,5],[2,8,5,10,3],'cD:',label='upper')        
plt.title('Scatterplot_1')  
plt.xlabel('X - axis');  plt.xlim(0,6)
plt.ylabel('Y - axis');  plt.ylim(1,11)
plt.legend(loc=2)

plt.subplot(2,1,2)
plt.plot([1,2,3,4,5],[8,4,6,3,9],'rs:',label='lower')        
plt.title('Scatterplot_2')  
plt.xlabel('X - axis');  plt.xlim(0,6)
plt.ylabel('Y - axis');  plt.ylim(1,11)
plt.legend(loc=2)
plt.tight_layout()
plt.show()     
 
```

```python
# 5. figure() & axes() function
x=[2,4,6,8,10]
y=[1,5,3,9,6]

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
graph=ax.plot(x,y,'kx-.')
ax.set_title("Scatterplot")
ax.set_xlabel("X - axis"); ax.set_ylabel("Y - axis")
ax.set_facecolor('yellow')
plt.show()
```
